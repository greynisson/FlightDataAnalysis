## Import packages:

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from datetime import datetime
```

### Define functions:

```python
def CalculatePunctuality(schedule, actual):
    schedule_mins = schedule.str[:2].astype('int32') * 60 + schedule.str[-2:].astype('int32')
    actual_mins = actual.str[:2].astype('int32') * 60 + actual.str[-2:].astype('int32')
    return actual_mins - schedule_mins
```
